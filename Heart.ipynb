```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
missing_values = ["?"]
df = pd.read_csv(r'hung.csv', names = ["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"], na_values = missing_values)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
m1 = df["chol"].median()
m2 = df["slope"].median()
m3 = df["ca"].median()
m4 = df["thal"].median()
m5 = df["trestbps"].median()
m6 = df["fbs"].median()
m7 = df["thalach"].median()
m8 = df["exang"].median()
m9 = df["restecg"].median()
```

```python
df["chol"].fillna(m1, inplace=True)
df["slope"].fillna(m2, inplace=True)
df["ca"].fillna(m3, inplace=True)
df["thal"].fillna(m4, inplace=True)
df["trestbps"].fillna(m5, inplace=True)
df["fbs"].fillna(m6, inplace=True)
df["thalach"].fillna(m7, inplace=True)
df["exang"].fillna(m8, inplace=True)
df["restecg"].fillna(m9, inplace=True)
```

```python
df
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
((df <= (Q1 - 1.5 * IQR)) | (df >= (Q3 + 1.5 * IQR))).sum()
```

```python
outlier_indices1 = []

list1 = ["chol","trestbps","oldpeak"]
for j in list1:
    for i in range(len(df)):
        if(df[j].iloc[i] <= (Q1[j] - 1.5 * IQR[j]) or (df[j].iloc[i] >= (Q3[j] + 1.5 * IQR[j]))):
            outlier_indices1.append(i)
    df = df.drop(outlier_indices1).reset_index().drop(['index'],axis=1) 
    outlier_indices1 = []
```

```python
X = df[["sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope","ca","thal","num"]]
y = df[["age"]]
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
X = pd.DataFrame(X)
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
linreg = LinearRegression()
```

```python
model = linreg.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
score = r2_score(y_test,predictions)
score
```
