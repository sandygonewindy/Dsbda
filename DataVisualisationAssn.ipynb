#Pie Chart

```python
vals = df['education'].value_counts()
labels = [i for i in df['education'].unique()]

data = []

for i in range(len(vals)):
  data.append(vals[i])

```

```python
plt.pie(vals, labels = labels, autopct='%.0f%%')
plt.show()
```

#Add histogram to box plot

```python
# creating a figure composed of two matplotlib.Axes objects (ax_box and ax_hist)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True)
 
# assigning a graph to each ax
sns.boxplot(df["Age"], ax=ax_box)
sns.histplot(data = df, x="Age", ax=ax_hist)
 
# Remove x axis name for the boxplot
ax_box.set(xlabel='')
plt.show()
```

#Add Scatter plot to box plot

```python
f, (ax_scatter, ax_box) = plt.subplots(2, sharex = True)

sns.stripplot(data = df, x = 'education-num', y = 'Age', ax = ax_scatter)
sns.boxplot(data = df, x = 'education-num', ax = ax_box)

ax_box.set(xlabel='')
plt.show()
```
